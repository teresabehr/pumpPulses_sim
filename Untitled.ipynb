```python
import numpy as np
import pandas as pd
# import scipy 
from scipy import special
import matplotlib.pyplot as plt
```

```python
degree = 3

p_monic = special.hermite(degree, monic=True)
p_monic
```

```python
p_monic(1)
```

```python
x = np.linspace(-5, 5, 400)

y = p_monic(x) * np.exp(-x**2)

plt.plot(x, y)

plt.title("third-order Hermit-Gaußian mode")

plt.xlabel("x")

plt.ylabel("H_3(x)")

plt.savefig("hg-third.png")

plt.show()
```
